Rol de Data Engineers (DEs)

```python
%sql
-- CATALOGO
GRANT ALL PRIVILEGES ON CATALOG catalog_au TO `DEs`;

-- PERMISOS ESQUEMAS
GRANT CREATE, USE SCHEMA ON SCHEMA catalog_au.bronze TO `DEs`;
GRANT CREATE, USE SCHEMA ON SCHEMA catalog_au.silver TO `DEs`;
GRANT CREATE, USE SCHEMA ON SCHEMA catalog_au.golden TO `DEs`;

-- PERMISOS TABLAS BRONZE
GRANT SELECT ON TABLE catalog_au.bronze.orders_raw TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.customers_raw TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.order_items_raw TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.products_raw TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.payments_raw TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.category_translation_raw TO `DEs`;

-- PERMISOS TABLA SILVER
GRANT SELECT ON TABLE catalog_au.silver.olist_orders_enriched TO `DEs`;

-- PERMISOS TABLA GOLDEN
GRANT SELECT ON TABLE catalog_au.golden.olist_sales_summary TO `DEs`;
```

```python
%sql
-- PERMISOS EXTERNAL LOCATION
GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-metastore`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-bronze`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-silver`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-golden`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-raw`
TO `DEs`;
```
