Revocar rol de Data Engineers (DEs)

```python

%sql
-- CATALOGO
REVOKE ALL PRIVILEGES ON CATALOG catalog_au FROM `DEs`;

-- PERMISOS ESQUEMAS
REVOKE CREATE, USE SCHEMA ON SCHEMA catalog_au.bronze FROM `DEs`;
REVOKE CREATE, USE SCHEMA ON SCHEMA catalog_au.silver FROM `DEs`;
REVOKE CREATE, USE SCHEMA ON SCHEMA catalog_au.golden FROM `DEs`;

-- PERMISOS TABLAS BRONZE
REVOKE SELECT ON TABLE catalog_au.bronze.orders_raw FROM `DEs`;
REVOKE SELECT ON TABLE catalog_au.bronze.customers_raw FROM `DEs`;
REVOKE SELECT ON TABLE catalog_au.bronze.order_items_raw FROM `DEs`;
REVOKE SELECT ON TABLE catalog_au.bronze.products_raw FROM `DEs`;
REVOKE SELECT ON TABLE catalog_au.bronze.payments_raw FROM `DEs`;
REVOKE SELECT ON TABLE catalog_au.bronze.category_translation_raw FROM `DEs`;

-- PERMISOS TABLAS SILVER
REVOKE SELECT ON TABLE catalog_au.silver.olist_orders_enriched FROM `DEs`;

-- PERMISOS TABLAS GOLDEN
REVOKE SELECT ON TABLE catalog_au.golden.olist_sales_summary FROM `DEs`;
```

```python
%sql
-- PERMISOS EXTERNAL LOCATION
REVOKE READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-metastore`
FROM `DEs`;

REVOKE READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-bronze`
FROM `DEs`;

REVOKE READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-silver`
FROM `DEs`;

REVOKE READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-golden`
FROM `DEs`;

REVOKE READ FILES, WRITE FILES
ON EXTERNAL LOCATION `exlt-raw`
FROM `DEs`;
```
